```python
import pandas as pd
import json
```

```python
s_name_2_json = {
    'TD_logopond': './logopond_metadata.json',
    'NL_hr_line': './vectorstock_hr_line.jsonl',
    'NL_artnivorastd': './test_vectorstock_artnivorastd.jsonl',
    'NL_brandingashell': './vectorstock_brandingashell.jsonl',
    'DL_brandingashell': './vectorstock_brandingashell.jsonl',
    'DL_iamguru': './vectorstock_iamguru.jsonl'
}
```

```python
xls = pd.ExcelFile('./VectorStock Data.xlsx')
```

```python
sheet_name = 'NL_hr_line'

def get_chosen_item_list(sheet_name):
    df = pd.read_excel(xls, sheet_name, names=['url'])
    url_list = df['url'].tolist()

    meta_data = []
    with open(s_name_2_json[sheet_name], 'r', encoding='utf8') as f:
        for line in f:
            meta_data.append(json.loads(line.strip()))

    uncollected_list = []
    collected_list = []
    for url in url_list:
        image_id = url.split('/')[-1].split('-')[-1]
        collected = False
        for item in meta_data:
            if image_id == item['image_id']:
                collected_list.append(item)
                collected = True
        if not collected:
            uncollected_list.append(url)
    
    return collected_list, uncollected_list

def get_chosen_item_list_logopond(sheet_name):
    df = pd.read_excel(xls, sheet_name, names=['url'])
    url_list = df['url'].tolist()
    meta_data = []
    with open(s_name_2_json[sheet_name], 'r', encoding='utf8') as f:
        meta_data = json.load(f)
    uncollected_list = []
    collected_list = []
    for url in url_list:
        collected = False
        for item in meta_data:
            if item['image_url'] == url:
                collected_list.append(item)
                collected = True
        if not collected:
            uncollected_list.append(url)
    
    return collected_list, uncollected_list
```

```python
json.dump(get_chosen_item_list_logopond('TD_logopond')[0], open('./chosen_logopond.json', 'w', encoding='utf8'), ensure_ascii=False)
```

```python
all_collected_metadata = []
for sheet_name in s_name_2_json.keys():
    all_collected_metadata.extend(get_chosen_item_list(sheet_name)[0])
```

```python
json.dump(all_collected_metadata, open('./chosen_vectorstock.json', 'w', encoding='utf8'), ensure_ascii=False)
```
