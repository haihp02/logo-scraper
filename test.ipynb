```python
import json

with open('./vectorstock_logoawesome.jsonl', 'r', encoding='utf8') as f:
    data = []
    for line in f:
        data.append(json.loads(line))
```

```python
tags = {}
for item in data:
    for tag in item['tags']:
        if tag not in tags:
            tags[tag] = 1
        else:
            tags[tag] += 1
tags = sorted(tags.items(), key=lambda x:x[1], reverse=True)
```

```python
tags
```

```python
not_vector_image = []
for item in data:
    if not item['title'].endswith('vector image'):
        not_vector_image.append(item)
```

```python
not_vector_image
```

```python
template_data, set_data, template_set_data = [], [], []
for item in data:
    if 'Icon' not in item['tags']:
        if 'Set' in item['tags'] and 'Template' in item['tags']:
            template_set_data.append(item)
        elif 'Set' in item['tags']:
            set_data.append(item)
        elif 'Template' in item['tags']:
            template_data.append(item)
```

```python
print(len(template_data))
print(len(set_data))
print(len(template_set_data))
```
